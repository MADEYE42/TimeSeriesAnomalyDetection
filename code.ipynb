```python
import cv2
import os
from glob import glob

# 👇 Set this to your dataset root folder (containing activity folders)
dataset_root = './Activities'  # Replace with the actual path in your Colab environment

# 👇 Output folder where annotated frames will be saved
output_root = './annotated_frames'
os.makedirs(output_root, exist_ok=True)

# 👇 Define your list of activity folders
activities = [
    'Falling', 'Jumping', 'Lying', 'Running', 'Shaking',
    'Sitting', 'Standing', 'Walking',
    'Walking Downstairs', 'Walking Upstairs'
]

# Normalize folder names
def normalize_activity_name(name):
    return name.replace(" ", "_").lower()

# Loop through each activity folder
for activity in activities:
    activity_path = os.path.join(dataset_root, activity)
    if not os.path.isdir(activity_path):
        print(f"Skipping (not a folder): {activity_path}")
        continue

    # Output folder for this activity
    activity_output_folder = os.path.join(output_root, activity)
    os.makedirs(activity_output_folder, exist_ok=True)

    # Process each video inside this folder
    video_files = glob(os.path.join(activity_path, '*.mp4'))  # or .avi, .mov
    for video_path in video_files:
        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"❌ Failed to open video: {video_path}")
            continue

        fps = cap.get(cv2.CAP_PROP_FPS)
        video_name = os.path.splitext(os.path.basename(video_path))[0]
        frame_count = 0
        saved_frame_count = 0

        while True:
            success, frame = cap.read()
            if not success:
                break

            current_time_ms = cap.get(cv2.CAP_PROP_POS_MSEC)
            if int(current_time_ms) % 1000 < 40:
                annotated_frame = frame.copy()
                cv2.putText(annotated_frame, activity.upper(), (10, 30),
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

                timestamp = int(current_time_ms / 1000)
                normalized_name = normalize_activity_name(activity)
                filename = f"{normalized_name}_{video_name}_{timestamp:04d}.jpg"
                filepath = os.path.join(activity_output_folder, filename)

                cv2.imwrite(filepath, annotated_frame)
                saved_frame_count += 1

            frame_count += 1

        cap.release()
        print(f"✅ {activity} | {os.path.basename(video_path)}: {saved_frame_count} frames saved")

print("\n🎉 Frame extraction and annotation complete!")
```
